```python
import numpy as np
import math 
import matplotlib.pyplot as plt
```

```python
def der(n):
    products = np.outer(n,n)
    squares =  0.5 * np.array([(np.triu(np.fliplr(products),i) - 
               np.triu(np.fliplr(products),i+1)).sum() 
                      for i in range(products.shape[0]-1,0,-1)])
    return squares - (1 + l) * n[1:] * n.sum()
N = 100
l = 0.3
h = 1e-2
steps = 2000
n = np.zeros((N,steps))
dn = np.array(N * [0])
n[0,:] = 1
for i in range(steps-1):
    n[1:,i+1] = n[1:,i] + h * der(n[:,i])
```

```python
for i in range(0,N):
    plt.plot(range(steps),n[i,:])
```
