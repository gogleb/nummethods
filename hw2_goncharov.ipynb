```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def true_fun(x):
    return np.sqrt(4 - np.power(x,2)) / 2
```

```python
def eq(x, y):
    return - x / (4 * y)
```

```python
def euler(dif, x, y, step):
    return y + eq(x,y) * step
```

```python
step_sizes = [21,201,2001]
for step_size in step_sizes:
    x = np.linspace(0,2,201)
    y = [1]
    for i, point in enumerate(x):
        y.append(euler(eq,point, y[i],x[1]-x[0]))
    plt.plot(x,y[:-1])
plt.legend(step_sizes);
```

```python
def runge_kutta(order, dif, x, y, step):
    if order == 2:
        y_half = y + step / 2 * dif(x,y)
        return y + step * dif(x + 0.5 * step, y_half)
    elif order == 4:
        k1 = dif(x,y)
        k2 = dif(x + step / 2, y + step / 2 * k1)
        k3 = dif(x + step / 2, y + step / 2 * k2)
        k4 = dif(x + step, y + step * k3)
        return y + 1 / 6 * step * (k1 + 2 * k2 + 2 * k3 + k4)
        
    else:
        raise Exception('Wrong order')
```

```python
step_sizes = [21,201,2001]
for step_size in step_sizes:
    x = np.linspace(0,2,201)
    y = [1]
    for i, point in enumerate(x):
        y.append(runge_kutta(2,eq,point, y[i],x[1]-x[0]))
    plt.plot(x,y[:-1])
plt.legend(step_sizes)
```

```python
step_sizes = [21,201,2001]
for step_size in step_sizes:
    x = np.linspace(0,2,201)
    y = [1]
    for i, point in enumerate(x):
        y.append(runge_kutta(4,eq,point, y[i],x[1]-x[0]))
    plt.plot(x,y[:-1])
plt.legend(step_sizes)
```

```python
x = np.linspace(0,2,21)
y_eu = [1]
y_r2 = [1]
y_r4 = [1]
for i, point in enumerate(x):
    y_eu.append(euler(eq,point, y_eu[i],x[1]-x[0]))
    y_r2.append(runge_kutta(2,eq,point, y_r2[i],x[1]-x[0]))
    y_r4.append(runge_kutta(4,eq,point, y_r4[i],x[1]-x[0]))
plt.scatter(x,y_eu[:-1],s=3,c='r')
plt.scatter(x,y_r2[:-1],s=3)
plt.scatter(x,y_r4[:-1],s=3)
print("Errors eu:", np.linalg.norm(true_fun(x) - y_eu[:-1]))
print("Errors r2:", np.linalg.norm(true_fun(x) - y_r2[:-1]))
print("Errors r4:", np.linalg.norm(true_fun(x) - y_r4[:-1]))
plt.plot(x,true_fun(x))
plt.legend(['Function', 'Euler', 'R2','R4',]);
```
